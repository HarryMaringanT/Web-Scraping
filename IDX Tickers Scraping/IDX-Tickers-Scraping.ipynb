```python
import bs4 as bs
import requests
import pandas as pd
 
resp = requests.get('https://id.wikipedia.org/wiki/Daftar_perusahaan_yang_tercatat_di_Bursa_Efek_Indonesia')
soup = bs.BeautifulSoup(resp.text, 'lxml')
table = soup.find('table', {'class': 'wikitable sortable'})
print(table)
```

```python
print(table.findAll('tr')[1:])
```

```python
tickers = []
company_names = []
listing_dates = []

for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text

        #second element is company name
        company_name = row.findAll('td')[1].text

        #thirth element is the sector
        listing_date = row.findAll('td')[2].text
        
        tickers.append(ticker)
        company_names.append(company_name)
        listing_dates.append(listing_date)
```

```python
print(tickers)
print(company_names)
print(listing_dates)
```

```python
new = [s.replace("\xa0", ":") for s in listing_dates]
print(new)
```

```python
listing_dates = [s.split("(",1)[1] for s in new]
listing_dates
```

```python
listing_dates = [s.replace(")\n", "") for s in listing_dates]
print(listing_dates)
```

```python
tickers = [s.replace("IDX: ", "") for s in tickers]
print(tickers)
```

```python
tickerdf = pd.DataFrame(tickers,columns=['Ticker'])
company_namedf = pd.DataFrame(company_names,columns=['Company Name'])
listing_datedf = pd.DataFrame(listing_dates,columns=['Listing Date'])
```

```python
df = pd.concat([tickerdf, company_namedf, listing_datedf], axis=1)
print(df)
```
